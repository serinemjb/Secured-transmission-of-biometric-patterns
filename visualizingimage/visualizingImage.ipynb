```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2 as cv
import ast
```

```python
with open('watermarked_LSB2.txt', 'r') as image:
    content = image.read()
```

```python
content = ast.literal_eval(content)
```

```python
watermarked = np.array(content, dtype=np.uint8)
```

```python
watermarked.reshape((256,288))
```
